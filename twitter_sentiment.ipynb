```python
#naming the columns
column_names=['target','id','date','flag','user','text']
twitter_data = pd.read_csv("sentiment140/training.1600000.processed.noemoticon.csv",names=column_names ,encoding="latin-1")
```

```python
#counting the number of missing values
twitter_data.isnull().sum()
```

```python
#checking the distribution of target column
twitter_data['target'].value_counts()
```

```python
#convert the target 4 to 1
twitter_data.replace({'target':{4:1}},inplace=True)
```

```python
twitter_data['target'].value_counts()
#0->negative 1->positive
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re

# Download stopwords (only needed once)
nltk.download('stopwords')

# Define stop words and stemmer
stop_words = set(stopwords.words('english'))
port_stem = PorterStemmer()
```

```python

port_stem = PorterStemmer()
def stemming(content):
    content = re.sub('[^a-zA-Z]', ' ', content)  # remove non-letters
    content = content.lower()                     # convert to lowercase
    words = content.split()                       # split into words
    words = [port_stem.stem(word) for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
twitter_data['stemmed_content']=twitter_data['text'].apply(stemming)
```

```python
print(twitter_data['stemmed_content'])
```

```python
#sepearting the data and label
X=twitter_data['stemmed_content'].values
Y=twitter_data['target'].values
```

```python
print(X)
print(Y)
```

```python
#train test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#coverting the textual data to numerical data
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```

```python
#model evaluation accuracy score on training data
X_train_prediction=model.predict(X_train)
training_data-accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy score on the training data :',training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data-accuracy = accuracy_score(Y_test,X_test_prediction)
```
